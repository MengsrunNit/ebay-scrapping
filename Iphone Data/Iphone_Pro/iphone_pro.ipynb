```python
import re
import pandas as pd

# Load data
iphone12pro = pd.read_csv('iphone_12_pro_only.csv')

# Step 1: Remove rows with '|' (multiple model options)
iphone12pro = iphone12pro[~iphone12pro['Title'].str.contains(r'\|', regex=True, na=False)]

# Step 2: Remove titles that list multiple models like "12 - 12 Pro" or "11 / 12 / 13"
pattern_multi = r'\b(?:iPhone\s*)?(?:[0-9]+)\b\s*[-/|]\s*\b(?:iPhone\s*)?(?:[0-9]+)(?:\s*(?:Pro|Pro\s*Max|Mini))?\b'
iphone12pro = iphone12pro[~iphone12pro['Title'].str.contains(pattern_multi, flags=re.IGNORECASE, na=False)]

# Step 3: Keep only “iPhone 12 Pro” — not 12, 12 Pro Max, or 12 Mini
iphone12pro = iphone12pro[
    iphone12pro['Title'].str.contains(r'\biPhone\s*12\s*Pro\b', flags=re.IGNORECASE, na=False)
    & ~iphone12pro['Title'].str.contains(r'\bPro\s*Max\b', flags=re.IGNORECASE, na=False)
    & ~iphone12pro['Title'].str.contains(r'\bMini\b', flags=re.IGNORECASE, na=False)
]

# Preview
iphone12pro
```
