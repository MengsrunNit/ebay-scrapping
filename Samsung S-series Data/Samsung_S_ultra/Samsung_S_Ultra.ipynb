```python
import re
import pandas as pd

s22ultra = pd.read_csv('Samsung_S22_Ultra.csv')

# Step 1: Remove rows with '|' (multiple model options)
s22ultra = s22ultra[~s22ultra['Title'].str.contains(r'\|', regex=True, na=False)]

# Step 2: Remove titles that list multiple models like "S22 / S22+" or "S22 - S22 Ultra"
pattern_multi = r'\bS?22\b\s*[-/|]\s*\bS?22\s*(?:Ultra|\+|Plus)?\b'
s22ultra = s22ultra[~s22ultra['Title'].str.contains(pattern_multi, flags=re.IGNORECASE, na=False)]

# Step 3: Keep only "Samsung Galaxy S22 Ultra"
s22ultra = s22ultra[
    s22ultra['Title'].str.contains(r'\b(Samsung\s*)?(Galaxy\s*)?S22\s*Ultra\b', flags=re.IGNORECASE, na=False)
    & ~s22ultra['Title'].str.contains(r'\bS22\s*(\+|Plus|Base|Regular)\b', flags=re.IGNORECASE, na=False)
]

# Preview the results
s22ultra
```
