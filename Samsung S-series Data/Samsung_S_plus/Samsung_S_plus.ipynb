```python
import re
import pandas as pd

s22plus = pd.read_csv('Samsung_S22_plus.csv')

# Step 1: Remove rows with '|' (multiple model options)
s22plus = s22plus[~s22plus['Title'].str.contains(r'\|', regex=True, na=False)]

# Step 2: Remove titles that list multiple models like "S22 / S22+" or "S22 - S22 Ultra"
pattern_multi = r'\bS?22\b\s*[-/|]\s*\bS?22\s*(?:Ultra|\+|Plus)?\b'
s22plus = s22plus[~s22plus['Title'].str.contains(pattern_multi, flags=re.IGNORECASE, na=False)]

# Step 3: Keep only "Samsung Galaxy S22 Plus" or "S22+"
s22plus = s22plus[
    s22plus['Title'].str.contains(r'\b(Samsung\s*)?(Galaxy\s*)?S22\s*(\+|Plus)\b', flags=re.IGNORECASE, na=False)
    & ~s22plus['Title'].str.contains(r'\bUltra\b', flags=re.IGNORECASE, na=False)
]

# Preview the results
s22plus
```

```python
import re
import pandas as pd

# Load the dataset
s25plus = pd.read_csv('Samsung_S25_plus.csv')

# Step 1: Remove rows with '|' (multiple model options)
s25plus = s25plus[~s25plus['Title'].str.contains(r'\|', regex=True, na=False)]

# Step 2: Remove titles that list multiple models like "S25 / S25+" or "S25 - S25 Ultra"
pattern_multi = r'\bS?25\b\s*[-/|]\s*\bS?25\s*(?:Ultra|\+|Plus)?\b'
s25plus = s25plus[~s25plus['Title'].str.contains(pattern_multi, flags=re.IGNORECASE, na=False)]

# Step 3: Keep only "Samsung Galaxy S25 Plus" or "S25+"
s25plus = s25plus[
    s25plus['Title'].str.contains(r'\b(Samsung\s*)?(Galaxy\s*)?S25\s*(\+|Plus)\b', flags=re.IGNORECASE, na=False)
    & ~s25plus['Title'].str.contains(r'\bUltra\b', flags=re.IGNORECASE, na=False)
    & ~s25plus['Title'].str.contains(r'\bBase|Regular\b', flags=re.IGNORECASE, na=False)
]

# Optional: remove duplicate listings
s25plus = s25plus.drop_duplicates(subset='Title', keep='first')

# Preview the results
s25plus
```
