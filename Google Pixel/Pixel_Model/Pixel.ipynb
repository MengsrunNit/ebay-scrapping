```python
import re
import pandas as pd

data = pd.read_csv('pixel_6_only.csv')

# Step 1: Remove rows with '|' (multiple model options)
data = data[~data['Title'].str.contains(r'\|', regex=True, na=False)]

# Step 2: Remove titles that list multiple models like "6 - 6 Pro" or "7a / 7 / 7 Pro"
pattern_multi = r'\b(?:[1-9][0-9]?)\b\s*[-/|]\s*\b(?:[1-9][0-9]?)(?:\s*(?:Pro|XL|a))?\b'
data = data[~data['Title'].str.contains(pattern_multi, flags=re.IGNORECASE, na=False)]

# Step 3: Keep only "Pixel 6" and NOT "6 Pro", "6 XL", "6a", etc.
data = data[
    data['Title'].str.contains(r'\bPixel\s*6\b', flags=re.IGNORECASE, na=False)
    & ~data['Title'].str.contains(r'\b6\s*(?:Pro|XL|a)\b', flags=re.IGNORECASE, na=False)
]

# Preview
data


```

```python
import re
import pandas as pd

data = pd.read_csv('pixel_9_pro_only.csv')

# 0) Drop any title that mentions "Fold" (e.g., "Fold", "Z Fold", etc.)
data = data[~data['Title'].str.contains(r'\bfold\b', flags=re.IGNORECASE, na=False)]

# 1) Remove rows with '|' (multiple model options)
data = data[~data['Title'].str.contains(r'\|', regex=True, na=False)]

# 2) Remove titles that list multiple models like "6 - 6 Pro" or "7a / 7 / 7 Pro"
pattern_multi = r'\b(?:[1-9][0-9]?)\b\s*[-/|]\s*\b(?:[1-9][0-9]?)(?:\s*(?:Pro|XL|a))?\b'
data = data[~data['Title'].str.contains(pattern_multi, flags=re.IGNORECASE, na=False)]

# 3) Keep only "Pixel 9 Pro" (and NOT 9 XL, 9a, etc.)
data = data[
    data['Title'].str.contains(r'\bPixel\s*9\s*Pro\b', flags=re.IGNORECASE, na=False)
    & ~data['Title'].str.contains(r'\b9\s*(?:XL|a)\b', flags=re.IGNORECASE, na=False)
]
data

# Optional: save
# data.to_csv('pixel_9_pro_only_clean.csv', index=False)
```
