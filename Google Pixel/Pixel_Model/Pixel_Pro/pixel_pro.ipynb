```python
import re
import pandas as pd

pixel6pro = pd.read_csv('pixel_6_pro_only.csv')

# Step 1: Remove rows with '|' (multiple model options)
pixel6pro = pixel6pro[~pixel6pro['Title'].str.contains(r'\|', regex=True, na=False)]

# Step 2: Remove titles that list multiple models like "6 - 6 Pro" or "7a / 7 / 7 Pro"
pattern_multi = r'\b(?:[1-9][0-9]?)\b\s*[-/|]\s*\b(?:[1-9][0-9]?)(?:\s*(?:Pro|XL|a))?\b'
pixel6pro = pixel6pro[~pixel6pro['Title'].str.contains(pattern_multi, flags=re.IGNORECASE, na=False)]

# Step 3: Keep only "Pixel 6 Pro" (not 6, 6 XL, 6a, etc.)
pixel6pro = pixel6pro[
    pixel6pro['Title'].str.contains(r'\bPixel\s*6\s*Pro\b', flags=re.IGNORECASE, na=False)
    & ~pixel6pro['Title'].str.contains(r'\b6\s*(?:XL|a)\b', flags=re.IGNORECASE, na=False)
]

# Preview
pixel6pro
```

```python
import re
import pandas as pd

pixel10pro = pd.read_csv('pixel_10_pro_only.csv')

# Step 1: Remove rows with '|' (multiple model options)
pixel10pro = pixel10pro[~pixel10pro['Title'].str.contains(r'\|', regex=True, na=False)]

# Step 2: Remove titles that list multiple models like "10 - 10 Pro" or "9 / 10 / 10 Pro"
pattern_multi = r'\b(?:[1-9][0-9]?)\b\s*[-/|]\s*\b(?:[1-9][0-9]?)(?:\s*(?:Pro|XL|a|Fold))?\b'
pixel10pro = pixel10pro[~pixel10pro['Title'].str.contains(pattern_multi, flags=re.IGNORECASE, na=False)]

# Step 3: Keep only "Pixel 10 Pro"; exclude XL/a/Fold and especially "10 Pro XL"
pixel10pro = pixel10pro[
    # must contain "Pixel 10 Pro" and NOT immediately followed by XL
    pixel10pro['Title'].str.contains(r'\bPixel\s*10\s*Pro\b(?!\s*XL\b)', flags=re.IGNORECASE, na=False)
    # exclude any form of "10 Pro XL" (hyphen, spaces, etc.)
    & ~pixel10pro['Title'].str.contains(r'\b10\s*Pro\s*[-/ ]*\s*XL\b', flags=re.IGNORECASE, na=False)
    # extra guards
    & ~pixel10pro['Title'].str.contains(r'\b10\s*(?:XL|a|Fold)\b', flags=re.IGNORECASE, na=False)
    & ~pixel10pro['Title'].str.contains(r'\bFold\b', flags=re.IGNORECASE, na=False)
]

# Preview
pixel10pro

```
